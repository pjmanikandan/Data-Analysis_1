# 3-Find the reason for non placement from the dataset

```python
nonp=dataset[dataset['status'] == 'Not Placed']
```

```python
nonp
```

```python
def quanQual(nonp):
    quan=[]
    qual=[]
    for columnName in nonp.columns:

        if(nonp[columnName].dtypes=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(nonp)
```

```python
quan
```

```python
qual
```

```python
 nonplaced=pd.DataFrame(index=["Mean","Median","Mode"],columns=quan)
for colname in quan:
    nonplaced[colname]["Mean"]=nonp[colname].mean()
    nonplaced[colname]["Median"]=nonp[colname].median()
    nonplaced[colname]["Mode"]=nonp[colname].mode()[0]
```

# 10-What is the probability Density Function of the salary range from 700000 to 900000

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'Green'},color='blue')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    sample=dataset
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f , Standard Deviation=%.3f' % (sample_mean, sample_std))
    dist=norm(sample_mean,sample_std)
    values=[value for value  in range (startrange,endrange)]
    probabilities=[dist.pdf(values) for values in values] 
    prob=sum(probabilities)
    print ("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
#with non placed student
get_pdf_probability(dataset["salary"],700000,900000)
```

```python
#placed student only
get_pdf_probability(plac["salary"],700000,900000)
```
